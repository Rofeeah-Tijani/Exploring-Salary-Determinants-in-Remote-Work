```python
import pandas as pd
```

**Display Of Variables In The Dataset**

```python
file_path = r"C:\Users\USER\Documents\Remote.xlsx"
df = pd.read_excel(file_path)
display(df.head())
```

**Descriptive Statistics And The Graphical Representation Of The Numerical Variable**

```python
num_variable = ['Annual_Salary']
print(df[num_variable].describe())
import matplotlib.pyplot as plt
df[num_variable].hist(figsize=(10, 6))
plt.title('Histogram of The Numerical Variable', fontsize=15)
plt.show()
```

**One Hot Encoding**

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
categorical_cols = ['Company','Job_Title','Industry', 'Employment_Type', 'Experience_Level', 'Remote_Flexibility']
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_features = encoder.fit_transform(df[categorical_cols])
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_cols), index=df.index)
y = df['Annual_Salary']
numerical_cols = ['Annual_Salary']
X = df.drop(columns=numerical_cols)
X = pd.concat([X.drop(columns=categorical_cols), encoded_df], axis=1)
display(X.head())
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

**Assumptions Of PCA**

**1) 📊Standardization**

```python
from sklearn.preprocessing import StandardScaler
numerical_cols = ['Annual_Salary']
numerical_df = df[numerical_cols]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(numerical_df)
scaled_df = pd.DataFrame(scaled_features, columns=numerical_cols, index=numerical_df.index)
final_df = pd.concat([scaled_df, encoded_df], axis=1)
display(final_df.head())
print("Shape of final_df:", final_df.shape)
```

**2) 🎛Correlation Between Variables**

```python
import seaborn as sns
correlation_matrix = final_df.corr()
plt.figure(figsize=(19, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

**3) 📑Suffient Sample Size**

```python
num_observations = final_df.shape[0]
num_variables = final_df.shape[1]
print(f"\nNumber of observations: {num_observations}")
print(f"Number of variables: {num_variables}")

if num_observations >= 10 * num_variables:
    print("Sufficient sample size for PCA.")
else:
    print("Insufficient sample size for PCA.  Consider obtaining more data.")
```

**📝Principal Component Analysis**

```python
from sklearn.decomposition import PCA
pca = PCA()
df_pca = pca.fit_transform(final_df)

explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = explained_variance_ratio.cumsum()

num_components = 0
for i, variance in enumerate(cumulative_variance_ratio):
    if variance >= 0.95:
        num_components = i + 1
        break

print(f"\nNumber of components explaining 95% variance: {num_components}")
print("Explained Variance Ratio:\n", explained_variance_ratio)

df_pca = pd.DataFrame(data=df_pca[:,:num_components])
display(df_pca.head())
```

**Variance Ratio**

```python
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--')
plt.title('Explained Variance Ratio per Principal Component')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance Ratio')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cummulative Explained Variance Ratio')
plt.show()
```

**📑Features Contribution To PC**

```python
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(len(explained_variance_ratio))], index=final_df.columns)
display(loadings.head(10))

#Top Features Driving PC1 and PC2
top_pcs = loadings[['PC1', 'PC2']].sort_values(by='PC1', ascending=False, key=abs)
display(top_pcs.head(10))
display(top_pcs.tail(10))

#Correlation Of Variables with target variables
correlations = df_pca.corrwith(y)
display(correlations.head(10))

#Relationship Between PC1, PC2 and Salary
plt.figure(figsize=(8, 6))
plt.scatter(df_pca[0], y, label='PC1')
plt.scatter(df_pca[1], y, label='PC2')
plt.xlabel('Principal Component')
plt.ylabel('Annual Salary')
plt.legend()
plt.title('Correlation of Principal Components with Annual Salary')
plt.show()

print("Interpretation:")
print(f"Top Features driving PC1: {', '.join(top_pcs.head(5).index.tolist())}")
print(f"Top Features driving PC2: {', '.join(top_pcs.head(5).index.tolist())}")
print(f"Correlation of PC1 with salary: {correlations[0]:.3f}")
print(f"Correlation of PC2 with salary: {correlations[1]:.3f}")
```

**📃 PCA Result Visualization (Scree Plot, Biplot, And Scatter Plot)**

```python
#1. Scree Plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--')
plt.title('Scree Plot: Explained Variance Ratio per Principal Component')
plt.xlabel('Principal Component Number')
plt.ylabel('Explained Variance Ratio')
plt.show()

#2. Biplot
plt.figure(figsize=(10, 8))
for i, feature in enumerate(final_df.columns):
    plt.arrow(0, 0, loadings.loc[feature, 'PC1'], loadings.loc[feature, 'PC2'], color='b', alpha=0.5)
    plt.text(loadings.loc[feature, 'PC1'] * 1.15, loadings.loc[feature, 'PC2'] * 1.15, feature, color='b', ha='center', va='center')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Biplot: Principal Components vs. Original Features')
plt.grid(True)
plt.show()

#3. Scatter Plot (PC1 vs. PC2 colored by Annual_Salary)
plt.figure(figsize=(10, 6))
plt.scatter(df_pca[0], df_pca[1], c=y, cmap='viridis') 
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Scatter Plot: PC1 vs. PC2 colored by Annual Salary')
plt.colorbar(label='Annual Salary')
plt.show()
```
